   ### 주요 파라미터
   - 키를 기준으로 DataFrame의 로우를 합친다. SQL 등 관계형 데이터베이스의 join 연산과 동일함.
   - 주요 파라미터        
      . left, right : merge할 DataFrame 객체이름
      
      . how = 'inner', #left, right, outer
      
      . on = None, #merge의 기준이 되는 컬럼
      
      . left_on = None, #left DataFrame의 기준 컬럼
      
      . right_on = None, #right DataFrame의 기준 컬럼

```python
df_left = pd.DataFrame({'KEY': ['k0', 'k1', 'k2', 'k3'],
                     'A': ['a0', 'a1', 'a2', 'a3'],
                     'B': ['b0', 'b1', 'b2', 'b3']})
df_right = pd.DataFrame({'KEY': ['k2', 'k3', 'k4', 'k5'],
                     'C': ['c2', 'c3', 'c4', 'c5'],
                     'D': ['d2', 'd3', 'd4', 'd5']})
```

```python
df_left
```

```python
df_right
```

```python
# inner join
pd.merge(df_left,df_right)
```

```python
# left outer join
pd.merge(df_left,df_right,how='left')
```

```python
#right outer join
pd.merge(df_left,df_right,how='right')
```

```python
# fully outer join
pd.merge(df_left,df_right,how='outer')
```

```python
# 실습 아래 2개의 DataFrame을 K1과 K2 컬럼으로 outer merge
df_left2 = pd.DataFrame({'K1': ['k0', 'k1', 'k2', 'k3'],
                     'A': ['a0', 'a1', 'a2', 'a3'],
                     'B': ['b0', 'b1', 'b2', 'b3']})
df_right2 = pd.DataFrame({'K2': ['k2', 'k3', 'k4', 'k5'],
                     'C': ['c2', 'c3', 'c4', 'c5'],
                     'D': ['d2', 'd3', 'd4', 'd5']}) 
```

```python
# 실습
pd.merge(df_left2,df_right2,how='outer',left_on='K1',right_on='K2')
```

```python
# 실제 데이터 실습
#  data/NC Dinos.xlsx 파일에서 2013, 2014, 2015시즌 정보를 읽어서, 각각 NC13, NC14, NC15에 저장한 후에, 
# 3개년도 정보를 하나의 데이터프레임 NCAll에 저장하기
NC = pd.read_excel('data/NC Dinos.xlsx',sheet_name=None)
NC13 = NC['2013'].drop('Unnamed: 0',axis=1)
NC14 = NC['2014'].drop('Unnamed: 0',axis=1)
NC15 = NC['2015'].drop('Unnamed: 0',axis=1)
# NC13, NC14, NC15 = NC.values()
```

```python
NC13['연도']=2013
NC14['연도']=2014
NC15['연도']=2015
```

```python
pd.concat([NC13,NC14,NC15],ignore_index=True)
```

## 4-3. concat(axis = 1)
<img src="img/4강/concat1예제.jpg" alt="csv파일예제" style="width: 800px;"/>

```python
# 샘플 데이터 생성
df1 = pd.DataFrame([[1,10],[2,20], [3,30]], 
                           columns = list('AB'), index = ['aa','bb', 'cc'])
df2 = pd.DataFrame([[3,'xx'],[4,'yy']], 
                           columns = list('CD'), index = ['aa','bb'])
```

```python
# concat with axis = 1 수행
pd.concat([df1, df2], axis = 1, sort=True)
```

###  실습. 로우 인덱스는 '선수명'으로, 컬럼 인덱스는 '연도'로 구성하여, 
### NC 다이노스 선수들의 홈런 기록을 연도별로 볼 수 있도록 데이터프레임을 구성하시오. 
<img src="img/4강/concat예제NC.jpg" alt="NC선수예제" style="width: 200px;"/>

```python
hr_data = pd.merge(
    pd.merge(NC13[['선수명','홈런']],NC14[['선수명','홈런']],
             on='선수명',suffixes=['_13','_14'],how='outer')
,NC15[['선수명','홈런']],on='선수명',how='outer').fillna('-')
```

```python
hr_data.columns=['선수명', '홈런_13', '홈런_14', '홈런_15']
hr_data
```
