```python
def mod(m,n):
    m=int(m)
    n=int(n)
    if m<0|n<0:
        print("매개변수는 모두 양의 정수이어야 합니다.")
    return m%n
```

```python
def ch_bin(string) :
    rst = {}
    for i in string :
        rst[i] = rst.get(i, 0) +1
    return(rst)

def print_dic(dic):
    for k, v in dic.items():
        print(k,v)

def find_dic(dic,n):
    rs =[]
    for k, v in dic.items():
        if v==n: rs.append(k)
    return(rs)

def change_dic(dic):
    rs ={}
    for k, v in dic.items():
        rs.get(v,[]).append(k)
    return(rs)

dic = ch_bin('banannnna goood')
print_dic(dic)
print(find_dic(dic,3))
print(change_dic(dic))
```

```python
def change_dic(dic):
    rs ={}
    for k, v in dic.items():
        rs[v]=rs.get(v,[])
        rs[v].append(k)
    return(rs)

dic = ch_bin('banannnna goood')
print(dic)
print(change_dic(dic))
```

```python
def inv_dic(d):
    new = {}
    for x in d :
        if d[x] not in new :
            new[d[x]]=[x]
        else :
            new[d[x]].append(x)
    return(new)

dic = ch_bin('banannnna goood')
print(dic)
print(inv_dic(dic))
```
